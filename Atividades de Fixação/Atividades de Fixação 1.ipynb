# 1. Operadores:

## 1.1 - Aritméticos
Exercício: Peça ao usuário dois números inteiros. Calcule e exiba a soma, a
diferença, o produto, a divisão real, a divisão inteira e o resto da divisão
(módulo) entre eles.

```python
n1 = int(input("Digite o 1° número: "))
n2 = int(input("Digite o 2° número: "))

adicao = n1 + n2
subtracao = n1 - n2
multiplicacao = n1 * n2
divisao_real = n1 / n2
divisao_inteira = n1 // n2
modulo = n1 % n2
potenciacao = n1 ** n2

print("")
print(f"Adição: {n1} + {n2} = {adicao}")
print(f"Subtração: {n1} - {n2} = {subtracao}")
print(f"Multiplicação: {n1} * {n2} = {multiplicacao}")
print(f"Divisão Real: {n1} / {n2} = {divisao_real:.2f}")
print(f"Divisão Inteira: {n1} // {n2} = {divisao_inteira}")
print(f"Módulo: {n1} % {n2} = {modulo}")
print(f"Potenciação: {n1} ** {n2} = {potenciacao}")
```

## 1.2 - Atribuição
Exercício: Crie uma variável saldo com valor inicial de 1000. Utilize
operadores de atribuição composta (como +=, -=, *=) para simular três
operações: um depósito de 500, um saque de 200 e um rendimento que
dobra o saldo atual. Exiba o resultado final.

```python
valor = 1000
deposito = 500
saque = 200
rendimento = 2

valor += deposito
valor -= saque
valor *= rendimento

print(f"O valor final do saldo é de: {valor}")
```

## 1.3 - Comparação
Exercício: Receba dois valores numéricos e retorne uma série de
booleanos (True ou False) verificando se o primeiro é maior que o segundo,
se são iguais e se o primeiro é diferente do segundo.

```python
n1 = int(input("Digite o 1° número: "))
n2 = int(input("Digite o 2° número: "))

print("")
print(f"{n1} é > que {n2}? {n1 > n2}")
print(f"{n1} é = {n2}? {n1 == n2}")
print(f"{n1} é < que {n2}? {n1 < n2}")
```

## 1.4 - Lógicos
Exercício: Um aluno só é aprovado se tiver presenca maior ou igual a
75 E media maior ou igual a 7.0. Receba esses dois valores e verifique se o
aluno foi aprovado utilizando operadores lógicos.

```python
presenca = float(input("Digite a presença do aluno: "))
media = float(input("Digite a média do aluno: "))
print("")

if (presenca >= 75) and (media >= 7.0):
  print("O aluno foi aprovado!")
else:
    print("O aluno foi reprovado!")
```

## 1.5 - Identidade (is, is not)
Exercício: Crie duas listas com os mesmos elementos, por exemplo: a = [1,
2] e b = [1, 2]. Crie uma terceira variável c = a. Use o operador de identidade
para verificar se a é o mesmo objeto que b e se a é o mesmo objeto que c.

```python
a = [20, 30, 40]
b = [20, 30, 40]
c = a

print(f"A{a} é o mesmo objeto que B{b}? {a is b}")
print(f"A{a} é o mesmo objeto que C{c}? {a is c}")
print("")
print(f"A{a} não é o mesmo objeto que B{b}? {a is not b}")
print(f"A{a} não é o mesmo objeto que C{c}? {a is not c}")
```

## 1.6 - Associação (in, not in)
Exercício: Dada a string "Ciência de Dados", verifique se o
caractere "D" está presente na frase e se a palavra "Python" não está
presente.

```python
print(f"O caractere 'D' está presente na frase 'Ciência de Dados'? {"D" in "Ciência de Dados"}")
print(f"O caractere 'D' não está presente na palavra 'Python'? {"D" not in "Python"}")
```

# 2. Estrutura Condicional (IF, ELIF, ELSE):
Exercício: Desenvolva um programa que receba a idade de um nadador e
classifique-o em uma das seguintes categorias:
• Infantil: 5 a 12 anos;
• Juvenil: 13 a 17 anos;
• Adulto: 18 anos ou mais;
• Inválido: Menor que 5 anos.

```python
idd = int(input("Digite sua idade: "))
if (idd >= 18):
  print("Adulto!")
elif (13 <= idd <= 17):
  print("Juvenil!")
elif (5 <= idd <= 12):
  print("Infantil!")
else:
  print("Inválido!")
```

# 3. Estrutura de Repetição (FOR):
Exercício: Peça um número inteiro ao usuário e exiba a tabuada de
multiplicação desse número de 1 a 10 utilizando o laço for e a
função range().

```python
n = int(input("Digite um número: "))
i = 0
for i in range(10):
    print(f"{n} X {i + 1} = {n * (i + 1)}")
```

# 4. Estrutura de Repetição (WHILE):
Exercício: Escreva um programa que peça ao usuário para digitar uma
senha. O programa deve continuar pedindo a senha (repetição) até que o
valor digitado seja igual a "python123". Quando acertar, exiba "Acesso
Permitido".

```python
senha = input("Digite uma senha: ")
while (senha != "python123"):
  senha = input("Digite uma senha: ")
print("")
print("Acesso permitido!")
```
