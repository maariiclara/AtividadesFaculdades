# Case: Você recebeu uma lista de transações de uma empresa. Seu trabalho é
criar um script que processe essas transações, verifique a integridade dos dados e
calcule o saldo final, mas com regras rígidas de segurança.
Requisitos do script:
1. Identidade e Atribuição: Comece com um saldo_inicial = 1000.0. Crie
uma variável checkpoint = saldo_inicial. Use o operador is para confirmar
que o checkpoint aponta para o mesmo local de memória.
2. Repetição e Associação: O programa deve pedir ao usuário o nome do
auditor. Verifique se o nome contém caracteres proibidos
como * ou # usando o operador in. Se tiver, peça para digitar novamente
(WHILE).
3. Processamento de Lote (FOR): O programa deve processar exatamente 4
transações. Para cada transação, o usuário insere o valor (positivo para
crédito, negativo para débito).
4. Condicionais e Lógica:

a. Se o valor for maior que 500.0, exiba um alerta: "Transação
de alto valor!".
b. Se o valor for negativo e o saldo atual for insuficiente para
cobri-lo (operadores lógicos), a transação deve ser ignorada e
uma mensagem de erro exibida.

5. Resultado Final: Exiba o saldo final e verifique se o saldo_final ainda é o
mesmo objeto que o checkpoint (não será, pois números são imutáveis em
Python e o valor mudou).

```python
saldo_inicial = 1000
checkpoint = saldo_inicial

print(f"A variável 'saldo_inicial' aponta para o mesmo local de memória de 'checkpoint'? {saldo_inicial is checkpoint}")
print("")

while True:
    nome = input("Digite o nome do auditor: ")

    for letra in nome:
        if letra in "!@#$%&*":
            print("--- Não use caracter especial como: ! @ # $ % & * ---")
            break
    else:
      break

transacao = 0
total_transacoes = 4
deposito = 0
saque = 0

print("")

for transacao in range(4):
    valor = int(input("Digite um valor para realizar uma transação: "))
    if (valor > 0):
        deposito += valor
        if (valor > 500):
            print("- Transação de alto valor! -")
        print("--- Depósito efetuado! ---")
    elif (valor < 0):
        if (saldo_inicial + valor >= 0): # EXEMPLO: 1000 + (-200) = 800 --- isso pode
            saque -= valor
            print("--- Saque efetuado! ---")
        else:
            print("--- Saldo menor, essa transação será ignorada! ---")
    else:
        print("--- Algo deu erro! Tente novamente ---")

saldo_inicial += deposito
saldo_inicial -= saque

print("")
print(f"Saldo atual de {saldo_inicial} reais")
print("")
print(f"A variável 'saldo_inicial' aponta para o mesmo local de memória de 'checkpoint'? {saldo_inicial is checkpoint}")
print("")
```
