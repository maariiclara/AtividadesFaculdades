# A Crise da Startup TechInvest

Imaginem que vocês acabaram de ser contratados pela TechInvest, uma startup que quer democratizar o acesso a investimentos. A empresa recebeu um aporte milionário, mas tem um problema grave: os analistas não estão dando conta de atender as milhares de pessoas que baixaram o App.

O CEO da empresa descobriu que vocês estão estudando Lógica com Python e deu uma missão: construir o 'cérebro' do App antes do final do dia. Vocês não podem usar bancos de dados complexos ainda. O que vocês têm é o poder da Lógica.

A Missão: Criar o "Prime" (O Primeiro Robô-Analista)
O objetivo do programa de vocês é ser o primeiro contato do cliente. Ele deve ser capaz de:
### ✓ Entender o Momento: Saber se o cliente está pronto para investir ou se precisa organizar as finanças primeiro.
### ✓ Classificar o DNA: Identificar se o cliente é medroso, cauteloso ou corajoso.
### ✓ Prever o Futuro: Mostrar para o cliente onde ele pode chegar se tiver disciplina.

### Fase 1 - A Entrevista (Entrada de Dados):
✓ Solicite ao cliente o nome, a renda, o gasto mensal e o seu "nível de coragem" (1 a 10).

### Fase 2 - O Diagnóstico (Operadores e Condicionais):
✓ Se o gasto for maior que a renda, o robô deve dar um alerta de "Emergência Financeira".
✓ Se sobrar dinheiro, o robô deve calcular quanto falta para ele ter uma reserva de segurança (6x o gasto mensal).

### Fase 3 - A Árvore de Decisão (Condicionais Compostas):
✓ Se coragem < 4: Tesouro Direto.
✓ Se coragem entre 4 e 7: Fundos Imobiliários.
✓ Se coragem > 7: Ações de Tecnologia.

### Fase 4 - A Máquina do Tempo (Repetição):
✓ O robô deve perguntar: "Quantos anos você pretende investir?".

### sem while true

```python
print("------- A Solução da Crise Startup TechInvest -------")
print("")
print("Seja Bem-Vindo(a) ao Sistema ClaraTech - Simulador de Investimentos")
print("")
print("-- Fase 1 - Entrevista --")
print("")

nome = input("Qual o seu nome? ")
renda = float(input("Qual a sua renda mensal? "))
gasto_mensal = float(input("Qual o seu gasto mensal? "))
ano_investimento = int(input("Quantos anos você pretende investir? "))
nivel_coragem = int(input("Qual o seu nível de coragem (1 a 10)? "))

# nível de coragem n pode ser maior q 10 e menor q 1
while (nivel_coragem > 10) or (nivel_coragem < 1):
  print("- Erro, digite de 1 a 10! -")
  nivel_coragem = int(input("Digite seu nível de coragem (1 a 10): "))

print("")
print("-- Fase 2 - Resultado do Perfil --")
print("")

print(f"{nome}, de acordo com as suas resposta obtemos esses resultados:")
print("")

if (gasto_mensal > renda):
    print("**********************************************************************************************************")
    print("* Seu diagnóstico é de EMERGÊNCIA FINANCEIRA! Cuidado você está gastando mais do que você recebe!")
    print("* Infelizmente, nesse caso não tem como você realizar nenhum tipo de investimento! :(")
    print("**********************************************************************************************************")
    print("")
    print("DICA: Reveja seus gastos e faça alguns cortes para equilibrar seu orçamento.")
    print("")
    print("------- Boa sorte na sua jornada de investimentos! ;) -------")
else:
    sobra_mes = renda - gasto_mensal

    if nivel_coragem > 7:
        print("Você deve investir em: Ações de Tecnologia!")
        investimento = 0.7 * sobra_mes # parte q sobra da renda e vai para o investimento
        rendimento = 0.012
    elif 4 <= nivel_coragem <= 7:
        print("Você deve investir em: Fundos Imobiliários!")
        investimento = 0.5 * sobra_mes
        rendimento = 0.008
    else:
        print("Você deve investir em: Tesouro Direto!")
        investimento = 0.3 * sobra_mes
        rendimento = 0.005

    reserva_seguranca = gasto_mensal * 6
    sobra_reserva = sobra_mes - investimento # já contabiliza oq sobrou do gasto mensal e do investimento para ir para reserva
    falta_para_reserva = reserva_seguranca - sobra_reserva

    # juros compostos
    meses = ano_investimento * 12
    total = 0
    i = 0
    for i in range(meses):
        total += investimento
        total = total * (1 + rendimento)

    # rendimento
    total_investido = investimento * meses
    lucro = total - total_investido

    print("")
    print("**********************************************************************************************************")
    print(f"* Quanto investir por mês: {investimento:.2f} reais -> {(investimento / sobra_mes) * 100}% do seu orçamento disponível (renda - gasto)")
    print(f"* Por quantos anos: {ano_investimento} anos")
    print(f"* Quanto acumulou (investimento + lucro): {total:.2f} reais")
    print(f"* Qual foi a taxa de rendimento mensal usada: {rendimento * 100}% -> valor calculado com base em juros compostos")
    print(f"* Quanto foi o rendimento (lucro): {lucro:.2f} reais")
    print("**********************************************************************************************************")
    print("")
    print(f"DICA: Tenha uma reserva de segurança (6x o gasto mensal) e falta {falta_para_reserva} reais para você ter a sua!")
    print("")
    print("------- Boa sorte na sua jornada de investimentos! ;) -------")
```

```python
print("------- A Solução da Crise Startup TechInvest -------")
print("")
print("Seja Bem-Vindo(a) ao Sistema ClaraTech - Simulador de Investimentos")
print("")
print("-- Fase 1 - Entrevista --")
print("")

nome = input("Qual o seu nome? ")
renda = float(input("Qual a sua renda mensal? "))
gasto_mensal = float(input("Qual o seu gasto mensal? "))
ano_investimento = int(input("Quantos anos você pretende investir? "))
nivel_coragem = int(input("Qual o seu nível de coragem (1 a 10)? "))

# nível de coragem não pode ser maior que 10 e menor que 1
while (nivel_coragem > 10) or (nivel_coragem < 1):
    print("- Erro, digite de 1 a 10! -")
    nivel_coragem = int(input("Digite seu nível de coragem (1 a 10): "))

print("")
print("-- Fase 2 - Resultado do Perfil --")
print("")

print(f"{nome}, de acordo com as suas respostas obtemos esses resultados:")
print("")

if gasto_mensal > renda:
    print("**********************************************************************************************************")
    print("* Seu diagnóstico é de EMERGÊNCIA FINANCEIRA! Cuidado você está gastando mais do que você recebe!")
    print("* Infelizmente, nesse caso não tem como você realizar nenhum tipo de investimento! :(")
    print("**********************************************************************************************************")
    print("")
    print("DICA: Reveja seus gastos e faça alguns cortes para equilibrar seu orçamento.")
    print("")
    print("------- Boa sorte na sua jornada de investimentos! ;) -------")
else:
    sobra_mes = renda - gasto_mensal

    if nivel_coragem > 7:
        print("Você deve investir em: Ações de Tecnologia!")
        investimento = 0.7 * sobra_mes  # parte que sobra da renda e vai para o investimento
        rendimento = 0.012
    elif 4 <= nivel_coragem <= 7:
        print("Você deve investir em: Fundos Imobiliários!")
        investimento = 0.5 * sobra_mes
        rendimento = 0.008
    else:
        print("Você deve investir em: Tesouro Direto!")
        investimento = 0.3 * sobra_mes
        rendimento = 0.005

    reserva_seguranca = gasto_mensal * 6
    sobra_reserva = sobra_mes - investimento  # já contabiliza o que sobrou do gasto mensal e do investimento para ir para reserva
    falta_para_reserva = reserva_seguranca - sobra_reserva

    # cálculo com juros simples
    meses = ano_investimento * 12
    total_capital = investimento * meses  # soma do que você investiu
    lucro_simples = investimento * rendimento * meses  # juros simples: sempre sobre o investimento mensal
    total_juros_simples = total_capital + lucro_simples

    print("")
    print("**********************************************************************************************************")
    print(f"* Quanto investir por mês: {investimento:.2f} reais -> {(investimento / sobra_mes) * 100:.2f}% do seu orçamento disponível (renda - gasto)")
    print(f"* Por quantos anos: {ano_investimento} anos")
    print(f"* Quanto acumulou (investimento + lucro) com juros simples: {total_juros_simples:.2f} reais")
    print(f"* Qual foi a taxa de rendimento mensal usada: {rendimento * 100:.2f}% -> valor calculado com base em juros simples")
    print(f"* Quanto foi o rendimento (lucro) com juros simples: {lucro_simples:.2f} reais")
    print("**********************************************************************************************************")
    print("")
    print(f"DICA: Tenha uma reserva de segurança (6x o gasto mensal) e falta {falta_para_reserva:.2f} reais para você ter a sua!")
    print("")
    print("------- Boa sorte na sua jornada de investimentos! ;) -------")
```
