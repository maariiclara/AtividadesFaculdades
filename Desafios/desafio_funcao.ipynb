# Desafio: Pipeline de Auditoria para Ciência de Dados
Contexto: Você atua como Cientista de Dados em uma rede varejista. A empresa
recebe volumes de vendas de diversas filiais, mas o sistema de coleta é instável,
gerando "ruídos" (valores negativos ou erros de digitação acima de R$ 10.000,00).
Sua missão é desenvolver uma ferramenta que receba esses dados, limpe-os e
gere um relatório de desempenho nominal para a sua gerência.
Requisitos Técnicos (Obrigatórios): Para este projeto, você deve aplicar:

1. Funções: Modularização da lógica de limpeza, cálculo e classificação.
2. Estrutura de Repetição (while): Para entrada dinâmica de dados.
3. Estrutura de Repetição (for): Para processar e listar as vendas validadas.
4. Listas e Tuplas: Armazenamento dos dados e retorno de múltiplas
métricas.
5. Condicionais (if/elif/else): Filtragem de erros e lógica de status de
performance.
6. Operadores: Cálculos e formatação de saída.
Especificações do Fluxo de Trabalho:


1. Identificação e Coleta:
• O programa deve iniciar solicitando o nome da Filial.
• Em seguida, deve abrir um ciclo de coleta de valores de vendas. O usuário
pode inserir quantos valores desejar.
• Critério de Parada: O loop de entrada deve ser encerrado quando o
usuário digitar o valor 0 (ZERO).

2. Tratamento de Dados (Função “limpar_vendas”):
• Crie uma função que receba a lista de vendas da filial.
• A função deve retornar uma nova lista contendo apenas valores que sejam
maiores que 0 e menores que 10.000.

3. Processamento Estatístico (Função “analisar_dados”):
• Crie uma função que receba a lista limpa.
• Ela deve calcular o Faturamento Total e a Média das Vendas.
• Retorno: A função deve obrigatoriamente retornar uma Tupla no formato
(total, media).

4. Classificação de Performance (Função “definir_status”):
• Crie uma função que receba a média calculada e retorne uma string
conforme os critérios:
o Média > 5000: "Alta Performance"
o Média entre 2000 e 5000: "Performance Estável"
o Média < 2000: "Performance Crítica"

5. Relatório de Saída (Exibição):
• O programa deve imprimir um relatório formatado contendo:
o Cabeçalho com o nome da filial.
o O status de performance.
o O faturamento total e a média (formatados com duas casas
decimais).
o Uma listagem numerada de todas as vendas que foram
consideradas válidas, utilizando um loop (for).

```python
filial = input("Digite o nome da Filial: ")

vendas = []
valor = 1

while valor != 0:
    valor -= 1
    valor = float(input("Digite o valor da venda (0 para encerrar): R$ "))
    vendas.append(valor)

def limpar_vendas(vendas):
    vendas_limpas = []
    for valor in vendas:
        if valor > 0 and valor < 10000:
            vendas_limpas.append(valor)
    return vendas_limpas
vendas_limpas = limpar_vendas(vendas)

def analisar_dados(vendas_limpas):
    total = sum(vendas_limpas)
    media = total / len(vendas_limpas)
    return total, media
total, media = analisar_dados(vendas_limpas)

def definir_status(media):
    if media > 5000:
        return "Alta Performance"
    elif 2000 <= media <= 5000:
        return "Performance Estável"
    else:
        return "Performance Crítica"

print("")
print("------ Relatório de Saída ------")
print("")
print(f"Nome da Filial: {filial}")
# print(f"Vendas Originais: {vendas}")
print(f"Faturamento Total: {total:.2f}")
print(f"Média do Faturamento: {media:.2f}")
print(f"Classificação de Performance: {definir_status(media)}")
print("")
print("Lista de Vendas Válidas (maior que 0 e menor que 10.000):")
for i in vendas_limpas:
    print(f"Valor da Venda Válida: R$ {i}")
```
